```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame(pd.read_csv('Dataset.csv'))
```

```python
my_list=df.columns.tolist()
for col in my_list:
    df[col].hist()
    plt.show()
    
    
```

Mode of plot

```python
my_list=['suicides_no','population','suicides/100k pop']
for col in my_list:
    df[col] = pd.to_numeric(df[col])
    
df.mode(axis=0,numeric_only=False,dropna=True).loc[0]
```

 i) Describing various aspects of columns by first converting to numeric and then using describe function on variables that are numeric.


```python

for cols in my_list:
    print(df[cols].describe())
    
     
    
    
    
```

Box and Whiskers plot

```python
for col in my_list:
    df[col].plot(kind='box')
    plt.show()
```
